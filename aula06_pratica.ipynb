```python
produto_a = [24.1, 29.2, 25.7, 26.5]
produto_b = [26.7, 31.8, 37.4, 28.1]
produto_c = [25.8, 30.0, 29.8, 28.3]

print('Media Produto A: ', np.mean(produto_a))
print('Media Produto B: ', np.mean(produto_b))
print('Media Produto C: ', np.mean(produto_c))

print('Desvio Produto A: ', np.std(produto_a, ddof=1))
print('Desvio Produto B: ', np.std(produto_b, ddof=1))
print('Desvio Produto C: ', np.std(produto_c, ddof=1))

plt.plot(produto_a, 'r--', label='produto A')
plt.plot(produto_b, 'bs--', label='produto B')
plt.plot(produto_c, 'g^--', label='produto C')

#plt.grid(True)
plt.title('Eficiencia em % no controlde de fungos agrícolas em 4 aplicações')
plt.legend()

plt.show()
```

***Resposta***

- 3 produtos foram aplicados mensalmente, durante 4 meses, em uma lavoura para testar eficácia no combate ao fungo, os resultados podem ser observados na tabela 1 a cima
- O produto B teve uma média de eficácia superior aos demais porem seu desvio padrão foi muito alto o que favorece o C que se mostrou mais estável 

2) Utilizando as tecnicas de web scraping vistas em sala de aula, crie uma aplicacao simples para capturar alguns dados, e gerar grafico. Comente os resultados

```python
# Deixe sua resposta aqui.

url = 'https://www.imdb.com/chart/top/?sort=rk,asc&mode=simple&page=1'
page = requests.get(url)

soup =bs4.BeautifulSoup(page.text, 'html.parser')

imdb_movies = []
imdb_rating = []

table= soup.find('table', class_='chart full-width')

for td in table.find_all(class_ = 'titleColumn'):
    imdb_movies.append(td.find('a').text)
    
for td in table.find_all(class_ = 'ratingColumn imdbRating'):
    imdb_rating.append(td.text.strip())
    
print(imdb_movies[:5])
print(imdb_rating[:5])

plt.barh(imdb_movies[:5],imdb_rating[:5],color='skyblue')
```
