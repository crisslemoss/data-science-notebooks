```
# inicialmete importar as bibliotecas
import pandas as pd
import numpy as np

```

```
# para rodar o Code tem que encontrar o valor de n
n=int(input('Digite a qtidade de valores: '))

x=np.random.randint(0,100,n)
y=np.arange(0,n)
z=np.linspace(0,20,21)
k=[2,np.nan,3,5,np.nan,10,np.nan,3,5,np.nan,0,0,0,3,4,300,1,1,0,0,2]
```

```
df=pd.DataFrame({'Col_A':x,'Col_B':y,'Col_C':z,'Col_D':k})
df
```

```
# neste Code encontre o erro

lista=[]
#for val in 'a b c d e f g h i j k l m n o p q r s t u':
for val in 'a b c d e f g h i j k l m n o p q r s t u':
    lista.append(val)

while(' ' in lista) : 
    lista.remove(' ')

df=pd.DataFrame({'Col_A':x,'Col_B':y,'Col_C':z,'Col_D':k},index=lista)
df
```
