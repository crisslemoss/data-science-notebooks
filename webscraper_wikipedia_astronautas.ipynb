```python
import re
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_astronautas_(1961%E2%80%931979)'
page = requests.get(url)
soup = bs4.BeautifulSoup(page.text, 'html.parser')

text = soup.find(class_ = 'wikitable')

rows = text.find_all('tr')
lista = []
for row in rows:
    tds = row.find_all('td')
    lista2 = []
    for td in tds:
        text2 = td.get_text()
        text2 = re.sub(r"[^0-9A-Za-zãõÃÕÂÊÔáéíóúçÇâêôà\-/ ]", '', text2)            
        if len(text2) == 0:
            lista2.append("-")
        else:
            lista2.append(str(text2).lstrip().rstrip())
    lista.append(lista2)

#print(lista)

#retira linha de titulo da tabela        
for item in lista:
    if len(item)==5:
        print(item[0], item[1], item[2])
        
print(lista)
```
